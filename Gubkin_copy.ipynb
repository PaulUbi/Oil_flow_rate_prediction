# Model Comparison with Crossvalidation

```python
#Define function for performance evaluation RMSE, MSE, MAE
def evaluate_(estimator, X = X_train, y = y_train):   
    y_pred =cross_val_predict(estimator, X, y, cv=5)
    mae=np.mean(mean_absolute_error(y_train,y_pred))
    print("MAE=",mae)
    mse=np.mean(mean_squared_error(y_train,y_pred))
    print("MSE=",mse)
    rmse=np.mean(np.sqrt(mean_squared_error(y_train,y_pred)))
    print("RMSE=",rmse)
    r2=np.mean(r2_score(y_train,y_pred,multioutput='raw_values'))
    print("R^2=",r2)
    
    #For Adjusted r^2 score
    N=len(y_train)
    k=7
    print("Adjusted R^2=",1-((1-r2)*((N-1)/(N-k-1))))
    
    print("MAPE=",np.mean(mean_absolute_percentage_error(y_train,y_pred)))
    print("EV=",np.mean(explained_variance_score(y_train,y_pred)))
    return [mae,mse,rmse]
```

```python
baseline_models=[ridge,rf,xgb]
hypertuned_models=[ridge_tuned,rf_tuned,xgb_tuned]
models_str=["Ridge Regression model","Random Forest model","XGBoost model"]
```

```python
#Model comparison for baseline models
c=0
metrics=[]
for model in baseline_models:
    print(models_str[c])
    metrics.append(evaluate_(model))
    c+=1
    print()
print('Metrics=',metrics)
```

```python
# Model comparison for tuned models
c=0
metrics_tuned=[]
for model in hypertuned_models:
    print(models_str[c])
    metrics_tuned.append(evaluate_(model))
    c+=1
    
    print()


print("Metrics of tuned models=",metrics_tuned)
```

```python
# Training speed
time_values=[]
import time
c=0
for model in hypertuned_models:
    t0=time.time()
    model.fit(X_train,y_train)
    print(f"For the {models_str[c]}")
    print("Training time:", time.time()-t0)
    time_values.append(time.time()-t0)
    c+=1
    print()
```

```python
# Regression plots for comparison
def error_compare_three(mods,X,y):
    f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=True)
    for mod, ax in ((mods[0], ax1),(mods[1], ax2),(mods[2], ax3)):
        predicted = cross_val_predict(mod[0], X, y, cv=5)
        ax.scatter(y, predicted, c='#F2BE2C')
        ax.set_title('Prediction Error for %s' % mod[1])
        ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2, c='blue')
        
    plt.xlabel('Measured')
    plt.ylabel('Predicted')
    plt.show()

models = np.array([(ridge_tuned,'Ridge'), (rf_tuned,'Random Forest'), (xgb_tuned,'XGBoost')])
```

```python
#Create prediction function
def make_prediction(ON_STREAM_HRS,BORE_GAS_VOL, AVG_CHOKE_SIZE_P,
                    AVG_WHP_P,AVG_WHT_P,AVG_DOWNHOLE_TEMPERATURE,BORE_WAT_VOL):
    data={
        "ON_STREAM_HRS":ON_STREAM_HRS,
        "AVG_DOWNHOLE_TEMPERATURE":AVG_DOWNHOLE_TEMPERATURE,
        "BORE_WAT_VOL":BORE_WAT_VOL,
        "AVG_CHOKE_SIZE_P":AVG_CHOKE_SIZE_P,
        'AVG_WHP_P':AVG_WHP_P,
        'AVG_WHT_P':AVG_WHT_P,
        'BORE_GAS_VOL':BORE_GAS_VOL
    }
    df=pd.DataFrame(data, index=[0])
    prediction = xgb_tuned.predict(df).round(2)[0]
    return f"Predicted well oil volume: {prediction}"
```
